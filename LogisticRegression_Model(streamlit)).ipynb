```python
import streamlit as st
import pickle
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
model= pickle.load(open('titanic_model','rb')) #brought log reg from another program we di before
model
```

```python
def main():
    st.title('Titanic Survival Prediction')
    st.image(r'titanic_sinking.jpg',caption='Sinking of Titanic : 15 April 1912 in North Atlanta Ocean',use_column_width=True)
    st.write("""## Would you have survived from this Titanic Disaster< """)
    #sidebar 
    st.sidebar.header('More Details:')
    st.sidebar.markdown('[ For more facts about Titanic here](https://www.telegraph.co.uk/travel/lists/titanic-fascinating-facts/#:~:text=1.,2.)')
    st.sidebar.markdown('[and here](https://titanicfacts.net/titanic-survivors/)')
    st.title("----    Check Your Survival Chances    ----")
    # User Interface of model
    age= st.slider("Enter Age :" ,1,75,30)
    fare= st.slider("Fare (in 1972 $) :",15,500,40)
    SibSp= st.selectbox("Number of Siblings or Spouses travelling with you: ", [0,1,2,3,4,5,6,7,8])
    Parch= st.selectbox("Number of Parent or Children travelling with you:", [0,1,2,3,4,5,6,7,8])

    sex= st.selectbox("Gender:" , ['Male','Female'])
    if (sex == "Male"):
        Sex=0
    else:
        Sex=1

    Pclass= st.selectbox("Select Passenger Class:",[1,2,3])

    boarded_location = st.selectbox("Select Location",['Cherbourg','Queenstown','Southampton'])
    Embarked_C, Embarked_Q, Embarked_S= 0,0,0
    if boarded_location == 'Queenstown' :
        Embarked_Q=1
    elif boarded_location == 'Southampton' :
        Embarked_S=1
    else:
        Embarked_C=1

    # Getting data and framing it into Dataframe
    data= {'Age':age,'Fare':fare,'SibSp':SibSp,'Parch':Parch,'Sex':Sex,'Pclass':Pclass,'Embarked_Q':Embarked_Q,'Embarked_S':Embarked_S,'Embarked_C':Embarked_C}
    df=pd.DataFrame(data,index=[0])
    return df
    
```

```python
## Predictions
if st.button("Predict"):
    result=model.predict(data)
    proba=model.predict_proba(data)
    if result[0] ==1:
        st.write("Congratulations!!, You probably would have made it!")
        st.image(r"lifeboat.jfif")
        st.write("Survival Probability Chances:'NO' : {}% 'Yes': {}% ".format(round((proba[0,0])*100,2),round((proba[0,1]*100,2))) )
    else:
        st.write("SORRY...!! You probably will die")
        st.image(r"Rip.jfif")
        st.write("Survival Probability Chances:'NO' : {}% 'Yes': {}% ".format(round((proba[0,0])*100,2),round((proba[0,1]*100,2))) )
```
